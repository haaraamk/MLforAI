```python
#팀원들의 이름, 취미, 국적으로 DF를 생성해보자.
data = {
    'name': ['이원준', '김동현', '강민지', '강하람'],
    'hobby': ['게임', '공연관람', '시트콤', '축구'],
    'nationality': ['대한민국' for _ in range(4)]
}

# DataFrame 생성
df = pd.DataFrame(data)

# 인덱스를 1부터 4까지 설정
df.index = [i for i in range(1, 5)]

# DataFrame 출력
df
```

```python
# 'name' 열을 새로운 인덱스로 설정
df = df.set_index('name')

# 새로운 인덱스로 설정한 데이터프레임 출력
print(df)
```

```python
# 데이터프레임을 CSV 파일로 저장
df.to_csv('파일명.csv', index=True, encoding='utf-8')

# 저장한 파일 확인
# 현재 디렉토리에 '파일명.csv' 파일이 생성됩니다.
```
